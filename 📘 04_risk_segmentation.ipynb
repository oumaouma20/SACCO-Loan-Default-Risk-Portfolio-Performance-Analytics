# Risk Segmentation
This notebook classifies SACCO members into risk categories
to support differentiated credit decisions.


```python
import pandas as pd

df = pd.read_csv("ugenyasacco_members.csv")
```

üè∑Ô∏è **Code Cell ‚Äî Create Risk Categories**

```python
def risk_category(pd_score):
    if pd_score < 0.2:
        return 'Low Risk'
    elif pd_score <= 0.5:
        return 'Medium Risk'
    else:
        return 'High Risk'

df['RiskCategory'] = df['DefaultRiskScore'].apply(risk_category)
```

**Risk Distribution**

```python
df['RiskCategory'].value_counts()
```

```python
df['RiskCategory'].value_counts().plot(kind='bar', title='Risk Distribution')
plt.show()

```

**Risk Concentration**

```python
df.groupby('RiskCategory')['LoanBalance'].sum()
```

### Key Insights
- High-risk members form a smaller group but hold a large share of loan balances.
- Risk segmentation enables targeted loan controls and monitoring.
- Medium-risk members represent the best opportunity for portfolio growth.
